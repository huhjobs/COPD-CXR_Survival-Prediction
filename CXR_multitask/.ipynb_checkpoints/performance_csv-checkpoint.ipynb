```python
import pickle
from glob import glob
import os
import pandas as pd
import yaml
```

```python
models = []
for f in os.listdir('./results/'):
    if '.' not in f:
        models.append(f)
```

```python
os.listdir(f'./results/{models[1]}/')
```

```python
performances = []
for model in models:
    for encoder in os.listdir(f'./results/{model}/'):
        if encoder == '.ipynb_checkpoints':
            pass
        exp = os.listdir(f'./results/{model}/{encoder}')
        if '.ipynb_checkpoints' in exp:
            exp.remove('.ipynb_checkpoints')
        for e in exp:
            try:
                with open(f'./results/{model}/{encoder}/{e}/performance.pickle','rb') as fp:
                    perf = pickle.load(fp)
            except FileNotFoundError:
                perf = {"C-index":None, "1y-AUC":None, "3y-AUC":None, "5y-AUC":None}
            performance = {'arch': model}
            performance.update({'encoder': encoder})
            performance.update({'exp': e})
            performance.update(perf)

            with open(f'./results/{model}/{encoder}/{e}/config.yaml') as fp:
                config = yaml.load(fp, Loader=yaml.FullLoader)
            config_lst = config.split('--')
            config = {}
            for conf in config_lst[1:]:
                key = conf.split(' ')[0]
                value = conf.split(' ')[1:]
                config.update({key: value})
            for conf in ["lr", "batch_size", "loss_weight"]:
                if conf in config.keys():
                    performance.update({conf:config[conf][0]})
            performances.append(performance)
```

```python
performances[0]
```

```python
results = pd.DataFrame(performances)
results.sort_values(by=['C-index'],ascending=False)
# results.sort_values(by=['1y-AUC'],ascending=False)
# results.sort_values(by=['3y-AUC'],ascending=False)
# results.sort_values(by=['5y-AUC'],ascending=False)
```
